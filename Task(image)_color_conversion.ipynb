```python
import cv2
import numpy as np
from PIL import Image
```

## Generic code:

```python
im_path = "/content/download1.png"


im_array = cv2.imread(im_path)

display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))

#In BGR format, define the range of red color 
def_min_range = (0, 0, 100)
def_max_range = (100, 100, 255)
# To get a binary mask of red_color pixels set the threshold of image
red_mask = cv2.inRange(im_array, def_min_range, def_max_range)
# Replace all red pixels with black pixels
im_array[red_mask != 0] = (0, 0, 0)

display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))
```
